## üîÅ Python Loop Exercises

These exercises will challenge your understanding of `for` loops, `while` loops, and control statements like `break` and `continue`.

### üöÄ Beginner Exercises (Basic Iteration)

1.  **Print Numbers:**
    * Use a **`for` loop** with the `range()` function to print all numbers from 1 to 10 (inclusive).

2.  **Countdown Timer:**
    * Use a **`while` loop** to print a countdown starting from 5 and ending at 1. Make sure the loop condition eventually becomes `False` to prevent an infinite loop.

3.  **Iterate a String:**
    * Start with `word = "DEVELOPER"`.
    * Use a **`for` loop** to iterate through the string and print each letter on a new line.

### üõ†Ô∏è Intermediate Exercises (Control Statements)

4.  **Skip Even Numbers (`continue`):**
    * Use a **`for` loop** that iterates through numbers from 1 to 10.
    * Use the **`continue`** statement to **skip** the iteration if the number is even.
    * Only print the odd numbers.

5.  **Stop at Target (`break`):**
    * Start with `data = [10, 20, 5, 30, 40]`.
    * Use a **`for` loop** to iterate through the list.
    * If the value **`5`** is encountered, print "Target found! Stopping loop." and use the **`break`** statement to exit the loop immediately.

6.  **Loop with `else`:**
    * Use a **`for` loop** to check the list `names = ["Alice", "Bob", "Charlie"]` for the name `"Zoe"`.
    * If `"Zoe"` is found, use `break`.
    * Use the **`else`** block associated with the loop to print "Zoe was not in the list."

### üß† Expert Exercises (Nesting and Advanced Logic)

7.  **Multiplication Table (Nested Loops):**
    * Use **nested `for` loops** to print the multiplication table for numbers 1 through 3.
    * *Outer loop:* Iterates for `i` from 1 to 3.
    * *Inner loop:* Iterates for `j` from 1 to 10.
    * Print the result of `i * j` in a formatted way (e.g., "1 x 1 = 1", "1 x 2 = 2", etc.).

8.  **Input Loop with Validation:**
    * Use a **`while True`** loop to repeatedly ask the user to enter a positive number.
    * If the user enters a number greater than 0, print the number and use **`break`** to exit the loop.
    * If the user enters a negative number or zero, print "Invalid input. Please try again." and use **`continue`** to prompt them again.

9.  **Password Guessing (Simulated):**
    * Set a secret password: `secret = "python123"`.
    * Use a **`for` loop** and the `range()` function to give the user **three attempts** to guess the password.
    * In each iteration, ask the user for input.
    * If the guess is correct, print "Access Granted!" and use **`break`**.
    * If the loop finishes without a correct guess, use the loop's **`else`** block to print "Access Denied. Too many attempts."
