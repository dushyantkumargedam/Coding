
## üöÄ Beginner Set Exercises (Fundamentals)

These focus on creation, adding/removing items, and understanding the unique nature of sets.

1.  **Set Creation and Duplicates:**
    * Create a set called `colors` with the following values: `"red", "blue", "green", "red", "yellow"`.
    * Print the set. What happened to the duplicate value?
    * Print the **length** of the resulting set.

2.  **Adding and Removing:**
    * Start with the set `prime_numbers = {2, 3, 5, 7}`.
    * Use the **`add()`** method to include the number `11`.
    * Use the **`remove()`** method to take out the number `2`.
    * Print the final set.

3.  **Safe Removal (`discard` vs. `remove`):**
    * Start with `letters = {"A", "B", "C"}`.
    * Use **`discard()`** to try and remove `"D"`. What happens?
    * Now, use **`remove()`** to try and remove `"E"`. What happens, and what error is raised? (This demonstrates the difference between the two methods).

## üõ†Ô∏è Intermediate Set Exercises (Logic and Operations)

These exercises involve using sets for filtering and performing mathematical set operations.

4.  **List to Set Conversion (Filtering Unique Values):**
    * Given the list of user IDs: `all_ids = [101, 105, 102, 105, 103, 101, 104]`.
    * Convert the list into a set called `unique_ids` to automatically filter out duplicates.
    * Convert `unique_ids` back into a list and print it.

5.  **Set Intersection (Common Elements):**
    * Define two sets:
        * `class_a = {"Alice", "Bob", "Charlie", "David"}`
        * `class_b = {"Charlie", "Ethan, "Fiona", "David"}`
    * Use the **`intersection()`** method or the **`&`** operator to find and print the names of students who are in **both** classes.

6.  **Set Difference (Unique Elements):**
    * Using `class_a` and `class_b` from the previous exercise, find the names of students who are in **`class_a` but NOT in `class_b`**.
    * Use the **`difference()`** method or the **`-`** operator.

## üß† Expert Set Exercises (Advanced Methods and Frozenset)

These exercises focus on more complex set methods and the concept of immutability.

7.  **Symmetric Difference (Exclusive Elements):**
    * Define two sets: `set1 = {1, 2, 3, 4, 5}` and `set2 = {4, 5, 6, 7, 8}`.
    * Use the **`symmetric_difference()`** method or the **`^`** operator to find and print all numbers that are present in **either** set, but **not** in both (the non-common elements).

8.  **Subset and Superset Check:**
    * Define three sets:
        * `large_set = {10, 20, 30, 40, 50}`
        * `small_set_a = {10, 30}`
        * `small_set_b = {10, 60}`
    * Use **`issubset()`** to check if `small_set_a` is a subset of `large_set`.
    * Use **`issuperset()`** to check if `large_set` is a superset of `small_set_b`. Print the Boolean results.

9.  **Frozenset and Dictionary Keys:**
    * **Frozensets** are hashable, meaning they can be used as **dictionary keys**.
    * Create a standard set `unhashable_set = {1, 2}`. Try to use it as a key in a dictionary (it will fail).
    * Convert `unhashable_set` into a **`frozenset`** called `hashable_set`.
    * Now, use `hashable_set` as the key in a dictionary where the value is `"Locked"`.
    * Print the dictionary to show the frozenset being used as a key.
