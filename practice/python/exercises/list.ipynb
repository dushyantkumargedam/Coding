## üöÄ Beginner List Exercises (Basic Operations)

These exercises focus on fundamental list creation, access, and modification.

1.  **List Creation and Access:**
    * Create a list called `planets` containing the names of the first four planets (Mercury, Venus, Earth, Mars).
    * Print the **length** of the list.
    * Print the planet at **index 2** (Earth).
    * Use **negative indexing** to print the last planet (Mars).

2.  **Adding and Removing:**
    * Start with the list `numbers = [10, 20, 30]`.
    * Use **`append()`** to add `40` to the end.
    * Use **`insert()`** to add `5` at the beginning (index 0).
    * Use **`remove()`** to take out the number `30`.
    * Print the final list.

3.  **Slicing a List:**
    * Given the list `alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g']`.
    * Create a new list called `middle` containing elements from index 2 up to, but not including, index 5 (`c`, `d`, `e`).
    * Create a new list called `start_end` containing the first two elements and the last two elements of the list (Hint: You'll need two slices and the `+` operator).


## üõ†Ô∏è Intermediate List Exercises (Logic and Iteration)

These exercises require looping, conditional logic, and using built-in methods.

4.  **Find the Maximum Value:**
    * Start with the list `scores = [85, 92, 78, 95, 88, 70]`.
    * Loop through the list using a **`for` loop** and find the **highest score** without using the built-in `max()` function.
    * Print the result.

5.  **Filtering with a Loop:**
    * Given the list `all_items = ["apple", "banana", "kiwi", "grape", "avocado", "orange"]`.
    * Create a new, empty list called `short_fruits`.
    * Use a **`for` loop** and an **`if` statement** to iterate through `all_items` and add only the fruits with a name length of **5 or less** to `short_fruits`.
    * Print `short_fruits`.

6.  **List Sorting and Reversal:**
    * Start with `mixed_names = ["Alice", "bob", "Charlie", "david"]`.
    * Sort the list **alphabetically** but make the sort **case-insensitive**.
    * After sorting, **reverse** the order of the list.
    * Print the final result.

## üß† Expert List Exercises (Advanced Techniques)

These exercises involve list comprehension, nested lists, and efficient data manipulation.

7.  **List Comprehension for Transformation:**
    * Start with the list `prices = [100, 250, 45, 1200]`.
    * Use a single-line **list comprehension** to create a new list called `discounted_prices` where every price has been reduced by **10%** (i.e., multiplied by 0.9).
    * Print `discounted_prices`.

8.  **Nested List Summation:**
    * Given the nested list `data = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]`.
    * Use **nested loops** (or a single list comprehension if you are feeling ambitious) to calculate the **sum of all numbers** within all the inner lists.
    * Print the grand total.

9.  **Removing Duplicates and Preserving Order:**
    * Given the list `data = ['A', 'B', 'A', 'C', 'B', 'D']`.
    * Create a new list called `unique_data` that contains only the unique items, and they must appear in the **same order** as they first appeared in the original list.
    * *(Hint: You will need a loop and the **`in`** keyword or a specialized data structure like a dictionary/set for tracking.)*

10. **The `split()` Challenge:**
    * Take the sentence `sentence = "Python lists are incredibly useful for data science tasks"`
    * Use the **`.split()`** string method to convert the sentence into a list of words.
    * Use a **list comprehension** to create a new list containing only the words whose **length is greater than 5**.
    * Print the final list of long words.
