````python
def type_agnostic_summation(arg):
  if type(arg) == list or type(arg) == tuple:
    return sum(arg)
  else:
    return 0


list_data = [1, 2, 3, 4]
sum_list = type_agnostic_summation(list_data)
print(f"List {list_data} sum: {sum_list}")

print(type(list_data))

# i having problem here i should get output as 10 but i am getting 0
# this happen only in notebook in  python it works
````
