

## üöÄ Beginner Exercises (Basic Conditionals)

These focus on using `if`, `else`, and simple comparison operators.

1.  **Simple Pass/Fail Check:**
    * Create a variable `score` and set it to a number (e.g., `65`).
    * Write an `if` statement that checks if the score is greater than or equal to `70`.
    * If `True`, print: "Passed the test."
    * Otherwise (using `else`), print: "Failed the test."

2.  **Even or Odd:**
    * Create a variable `number` and set it to an integer (e.g., `9`).
    * Write an `if` statement that uses the **Modulus Operator (`%`)** to check if the number is even (remainder is 0 when divided by 2).
    * Print whether the number is "Even" or "Odd".

3.  **Short-Hand If-Else:**
    * Create two variables, `a = 15` and `b = 20`.
    * Use the **short-hand `if-else`** syntax to assign the value "B is Larger" to a variable called `result` if `b` is greater than `a`, otherwise assign "A is Larger or Equal".
    * Print `result`.

## üõ†Ô∏è Intermediate Exercises (Elif and Logical Operators)

These exercises require using the `elif` keyword and combining conditions with `and`, `or`.

4.  **Grade Calculator (If-Elif-Else Chain):**
    * Create a variable `points` and set it to a value (e.g., `88`).
    * Write a complete `if-elif-else` chain to determine and print the letter grade:
        * `90` or above: "A"
        * `80` to `89`: "B"
        * `70` to `79`: "C"
        * `60` to `69`: "D"
        * Below `60`: "F"

5.  **Vacation Day Planner (`and` operator):**
    * Create variables `temperature = 28` and `is_raining = False`.
    * Write an `if` statement that uses the **`and`** operator to check **two conditions**:
        1.  The `temperature` is greater than `25`.
        2.  `is_raining` is `False`.
    * If both are `True`, print: "Perfect weather for the beach!" Otherwise, print: "Stay inside."

6.  **Weekend Check (`or` operator):**
    * Create a variable `day = "Sunday"`.
    * Write an `if` statement that uses the **`or`** operator to check if the `day` is "Saturday" **or** "Sunday".
    * Print "It's the weekend!" or "Back to work."

## üß† Expert Exercises (Nesting, Multiple Logic, and `pass`)

These exercises involve nesting, using `not`, and applying advanced flow control.

7.  **Ticket Price Logic (`and` and `not`):**
    * Create variables: `age = 17`, `is_student = True`.
    * Write an `if` statement to determine if a person qualifies for a **student discount**:
        * The person must be **under 18** OR a **student** (or both).
        * The person must **NOT** be 65 or older.
    * Use parentheses to group your conditions for clarity. If the criteria are met, print: "Eligible for Discount." Otherwise, print: "Standard Price."

8.  **Nested Conditional Access Control:**
    * Create variables: `is_logged_in = True`, `user_role = "admin"`.
    * Use **nested `if` statements** to check the following:
        * **Outer `if`**: Check if `is_logged_in` is `True`.
        * **Inner `if`**: If logged in, check if `user_role` is `"admin"`.
    * If both are true, print: "Access Granted to Admin Panel."
    * If logged in but **not** admin, print: "Access Denied."
    * If not logged in, print: "Please log in first."

9.  **Placeholder Logic (`pass` statement):**
    * Create a variable `product_status = "pending"`.
    * Write an `if-elif-else` chain to handle different statuses:
        * If `product_status` is `"available"`, print: "Ready for shipment."
        * If `product_status` is `"pending"`, use the **`pass`** statement (code to be written later).
        * If `product_status` is `"out_of_stock"`, print: "Order cancelled."
    * Execute the code and confirm that the "pending" condition results in no action.
