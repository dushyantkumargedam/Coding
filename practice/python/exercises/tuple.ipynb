## üöÄ Beginner Tuple Exercises (Fundamentals)

These focus on creation, accessing, and understanding immutability.

1.  **Creation and Access:**
    * Create a tuple called `months` containing the names of the first three months of the year (`January`, `February`, `March`).
    * Print the length of the tuple.
    * Access and print the month at **index 1**.
    * Use **negative indexing** to print the last month.

2.  **Single-Item Tuple:**
    * Create a variable `single_item` containing the string `"Data"` and ensure its type is a **tuple**.
    * Print the variable and use the **`type()`** function to prove it is a tuple.

3.  **Immutability Test:**
    * Start with `coordinates = (10, 20)`.
    * Attempt to change the first element to `15` using index assignment (`coordinates[0] = 15`).
    * What kind of **error** does Python raise? (This confirms the immutable property).

## üß† Expert Tuple Exercises (Unpacking and Nesting)

These exercises focus on unpacking and using tuples in advanced ways.

7.  **Basic Unpacking:**
    * Create a tuple `rgb_color = (255, 0, 128)`.
    * Unpack this tuple into three variables named `red`, `green`, and `blue` in a single line.
    * Print the value of `blue`.

8.  **Variable Assignment Unpacking (Asterisk `*`):**
    * Given `product_details = ("SKU001", "Laptop Pro", 1200.50, 5, "Electronics", "Warehouse A")`.
    * Unpack the details into variables such that:
        * `sku` gets the first item.
        * `price` gets the third item.
        * `rest_details` gets all the remaining items using the asterisk `*`.
    * Print the `rest_details` variable. What data type is it?

9.  **Nested Tuple Access:**
    * Given the nested tuple of records: `records = (("Alice", 30), ("Bob", 45), ("Charlie", 22))`.
    * Access and print the age of the second person ("Bob").
    * Use a **`for` loop** to iterate through the `records` and print a sentence for each person, like: "Alice is 30 years old."
