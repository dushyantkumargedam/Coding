
## üöÄ Beginner Dictionary Exercises (Fundamentals)

These focus on creation, basic access, and simple modification.

1.  **Creation and Access:**

      * Create a dictionary called `person` with the keys `"name"`, `"age"`, and `"city"`. Assign appropriate values to each key.
      * Use **square brackets `[]`** to print the person's name.
      * Use the **`.get()`** method to safely print the person's city.

2.  **Adding and Changing:**

      * Start with `inventory = {"apples": 10, "bananas": 15}`.
      * Change the value of `"apples"` to `12`.
      * **Add** a new item `"oranges"` with a count of `8`.
      * Print the final `inventory` dictionary.

3.  **Key Check and Length:**

      * Using the `inventory` dictionary from the previous exercise, use the **`in`** keyword to check if the key `"grapes"` exists. Print the Boolean result.
      * Use the **`len()`** function to print the total number of items in the dictionary.
